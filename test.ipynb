```python
import PM_tSNE
import numpy as np

X = np.load('/home/valentin/ownCloud/Personnel/Memoire/test_PM_2.0/MNIST_data.npy', allow_pickle=True)
y = np.load('/home/valentin/ownCloud/Personnel/Memoire/test_PM_2.0/MNIST_labels.npy', allow_pickle=True)
color = ['red','blue','green','orange','black','pink','yellow','brown','purple','grey']
c = np.asarray(color)[y.astype(int)[:]]
```

```python
tsne = PM_tSNE.PM_tSNE(n_iter=750, coeff=8, grid_meth='NGP')
Y = tsne.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 10, 10

plt.scatter(Y[:,0], Y[:,1], c=c[:], s=3)
```
